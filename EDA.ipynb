```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Style
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [12, 8]

# Dataset loading
df = pd.read_csv('truelayer_data_sc_test_data_set.csv')
```

## Exploratory Data Analysis

```python
# Dataset info
df.info()
```

```python
# Percentage of null values
round(df.isnull().sum() / df.shape[0] * 100, 1)
```

```python
# Categorical Columns values examples
cat_cols = df.select_dtypes(include=['object']).columns
for c in cat_cols:
    print(df[c].value_counts())
```

```python
# Status seems to have some values with different formatting
df.status.value_counts()
```

```python
# initiated = Initiaded
# failed = Failed
df.status = df.status.str.lower()
df.status.value_counts()
```

```python
# Assuming that status is the target column
sns.countplot(x='status', data=df, order=df['status'].value_counts().index)
plt.title('Status Distribution')
plt.xticks(rotation=45)
plt.show()
```

Based on truelayer docs - https://docs.truelayer.com/docs/payment-statuses - the terminal status are 'executed', 'settled', and 'failed'.
Since other status are not indicative of wheter a payment as failed yet or not, we will restrict our analysis to these 3 status - in our case we wonl have executed and failed.

```python
# Validation no other stage has a failure reason
df[~df.status.isin(['executed','failed'])].failure_reason.value_counts()
```
